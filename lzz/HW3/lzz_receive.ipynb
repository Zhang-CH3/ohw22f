```python
import cv2
import numpy as np
import serial
ser=serial.Serial("COM13",timeout=0.1)
light = cv2.imread("R.png",-1)
light = cv2.resize(light,(50,50))
while True:
    resp = ser.readline()
    loc = resp.decode()
    print(loc)
    img = np.zeros((800, 800, 3), np.uint8)
    if loc == "left":
        img[400:450,0:50] = light
        cv2.imshow("q",img)
    elif loc == "right":
        img[400:450,750:800] = light
        cv2.imshow("q",img)
    elif loc == "top":
        img[0:50,400:450] = light
        cv2.imshow("q",img)
    elif loc == "down":
        img[750:800,400:450] = light
        cv2.imshow("q",img)
    else:
        cv2.imshow("q",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cv2.destroyAllWindows()
```
